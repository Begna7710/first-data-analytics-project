```python
null_rows = df[df['StockItemID'].isnull() & df['Description'].isnull()]
```

```python
num_null_rows = len(null_rows)
print(num_null_rows)
```

```python
df = df[~(df['StockItemID'].isnull() & df['Description'].isnull())]
```

```python
print(df)
```

```python
df.shape
```

```python
df['TaxRate'] = df['TaxRate'].apply(lambda x: -x if x < 0 else x)
```

```python
df.loc[df['TaxRate']< 0, 'TaxRate'] = df.loc[df['TaxRate']< 0, 'TaxRate'] * -1
```

```python
df.describe()
```

```python
df.loc[df['TaxAmount'].isnull(), 'TaxAmount'] = (df['UnitPrice'] * df['Quantity'] * df['TaxRate']) / 100
```

```python
df.describe()
```

```python
df['ExtendedPrice'] = df['ExtendedPrice'].replace(-999, np.nan)
```

```python
df.loc[df['ExtendedPrice'].isnull(), 'ExtendedPrice'] = (df['UnitPrice'] * df['Quantity']) + df['TaxAmount']
```

```python
df.describe()
```

```python
from scipy.stats import zscore
```

```python
df['Quantity_zscore'] = zscore(df['Quantity'])
df['ExtendedPrice_zscore'] = zscore(df['ExtendedPrice'])
outliers = df[(df['Quantity_zscore'] >= 3) | (df['ExtendedPrice_zscore'] >= 3)]
print(outliers)
```

```python
df['Quantity_zscore'] = zscore(df['Quantity'])
quantity_outliers = df[df['Quantity_zscore'] >= 3]
print(quantity_outliers)
```

```python
df['ExtendedPrice_zscore'] = zscore(df['ExtendedPrice'])
extended_price_outliers = df[df['ExtendedPrice_zscore'] >= 3]
print(extended_price_outliers)
```
